Probability Of The Distribuction Table

```python

# random sampling and pass the final exam
# random  sample
random_sample = data.sample(5)
print(random_sample)

# count the student who pass the exam
pass_exam = (random_sample['pass_exam'] == 'Yes').sum()
print( "random student pass the final exam: ",pass_exam)

# probability of the pass in the sample number of 5 student
probability_of_pass = len(random_sample) / pass_exam
print("probability of pass the random sample: ",probability_of_pass)

#  probability distribuctin table
n = len(random_sample)
p = probability_of_pass

# Binomial PMF Function
def binom_pmf(k, n, p):
    from math import comb
    return comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# X values (0 to n)
x_values = np.arange(n + 1)

# PMF values
pmf_values = [binom_pmf(k, n, p) for k in x_values]

# Create distribution table
prob_dist_table = pd.DataFrame({
    "X (Number of Students Passing)": x_values,
    "P(X = x)": np.round(pmf_values, 4)
})

print("\nProbability Distribution Table:")
print(prob_dist_table)

# calculate a mean and veriance

mean = n * p
print("mean: ",mean)

variance = n * p * (1 - p)
print("variance: ",variance)
```
