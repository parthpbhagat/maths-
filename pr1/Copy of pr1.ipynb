```python
import pandas as pd
import numpy as np
import statistics
data = pd.read_csv("records.csv")
print(data)
```

```python

# calculated mean,median and mode
print("mean: ",data.household_income.mean())
print("median: ",data.household_income.median())
print("mode: ",data.household_income.mode())
```

```python
# measures of central tendancy range, variance and standard deveaction

range = data.household_income.max() - data.household_income.min()
print("range: ",range)

variance = data.household_income.var()
print("variance: ",variance)

standard_deviation = data.household_income.std()
print("standard deviation: ",standard_deviation)
```

```python
# distribuction
import matplotlib.pyplot as plt
plt.hist(data.household_income)
plt.show()
```

```python
skewness = data.household_income.skew()
print("skewness of household income: ",skewness)

kurtosis = data.household_income.kurtosis()
print("kurtosis of household income: ",kurtosis)
```
